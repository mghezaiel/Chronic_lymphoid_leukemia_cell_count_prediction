## Periodogramme

```python
fig,ax = plt.subplots(figsize = (7,7))
for signal in range(data.shape[0]):
    signal = data.filter([col for col in data.columns if "lambda" in col],axis = 1).values[signal,:]
    fs = 10 
    f, Pxx_den = scipy.signal.periodogram(signal, fs)
    ax.semilogy(f,Pxx_den)
    ax.set_xlabel("Frequency (Hz)",fontsize = 15)
    ax.set_ylabel("Power spectrum density (PSD)",fontsize = 15)
ax.vlines(ymin = 1e-22,ymax = max(Pxx_den),x = 2,ls ="dashed")
plt.show()
```

```python
fig.savefig("PSD.png")
```

```python
log_sum_PSD = []
for ids,signal in enumerate(range(data.shape[0])):
    signal = data.filter([col for col in data.columns if "lambda" in col],axis = 1).values[signal,:]
    fs = 10 
    f, Pxx_den = scipy.signal.periodogram(signal, fs)
    PSD_freq = [freq for freq in zip(f,Pxx_den)]
    PSD_signal = [freq[1] for freq in PSD_freq if freq[0]<=2]
    
    log_sum_PSD.append(np.log(sum(PSD_signal)))
    #plt.semilogy(f,Pxx_den)
```

```python
import plotly.tools as tls
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
f,ax = plt.subplots(3,1,figsize = (17,10))
data["log_sum_PSD"] = log_sum_PSD
data["spectrum_index"] = data.index
sns.scatterplot(data = data, x = "spectrum_index",y = "log_sum_PSD", hue = "patient_name",palette = "viridis",ax = ax[0])
ax[0].hlines(y = np.median(log_sum_PSD), 
         xmin = 0, xmax = len(log_sum_PSD), lw = 3, color = "r")
ax[0].legend('')
sns.scatterplot(data = data, x = "spectrum_index",y = "log_sum_PSD", hue = "patient_state",palette = "magma",ax = ax[1])
ax[1].hlines(y = np.median(log_sum_PSD), 
         xmin = 0, xmax = len(log_sum_PSD), lw = 3, color = "r")
sns.scatterplot(data = data, x = "spectrum_index",y = "log_sum_PSD", hue = "cell_type",palette = "mako", ax = ax[2])
ax[2].hlines(y = np.median(log_sum_PSD), 
         xmin = 0, xmax = len(log_sum_PSD), lw = 3, color = "r")

plt.tight_layout()

plt.show()
PSD_filtered_spectra = [1 if PSD>np.median(log_sum_PSD) else 0 for PSD in log_sum_PSD]
data["PSD_filtered"] = PSD_filtered_spectra
print(Counter(PSD_filtered_spectra))
```

```python
# Nombre de patient avant et aprés filtrage 
# Nombre de cellule B et TNK avant et aprés filtrage
# Nombre de patient sains et malades avant et aprés filtrage 
# Ratio B TNK avant et aprés filtrage 
f,ax = plt.subplots(2,2,figsize = (15,7))
ax = ax.flatten()
sns.countplot(x = data["PSD_filtered"].apply(lambda x: "Kept" if x==1 else "Filtered"), ax = ax[0],palette = "mako")
columns = ["patient_state","cell_type"]
for idc,column in enumerate(columns):
    spectra_counts_before_filtering = Counter(data[column])
    spectra_counts_before_filtering = pd.DataFrame({column:spectra_counts_before_filtering.keys(), 
                                               "counts":spectra_counts_before_filtering.values()})
    spectra_counts_after_filtering = Counter(data[data["PSD_filtered"]==1][column])
    spectra_counts_after_filtering = pd.DataFrame({column:spectra_counts_after_filtering.keys(), 
                                              "counts":spectra_counts_after_filtering.values()})
    spectra_counts = pd.concat([spectra_counts_before_filtering,spectra_counts_after_filtering],axis = 0)
    spectra_counts["filtering"] = ["Before filtering" if val<spectra_counts_before_filtering.shape[0]
                              else "After filtering" for val in range(spectra_counts.shape[0])]

    sns.barplot(data = spectra_counts, x = "filtering", y = "counts",hue = column, ax = ax[idc+1])
    ax[idc+1].legend_.set_title(None)

b_tnk_ratio = list()
b_all_cells_ratio = list()

for f in spectra_counts["filtering"].unique(): 
    curr_spectra_counts = spectra_counts[spectra_counts["filtering"]==f]
    b_tnk_ratio.append(curr_spectra_counts[curr_spectra_counts["cell_type"]=="B"]["counts"].values[0]/
                 curr_spectra_counts[curr_spectra_counts["cell_type"]=="TNK"]["counts"].values[0])
    b_all_cells_ratio.append(curr_spectra_counts[curr_spectra_counts["cell_type"]=="B"]["counts"].values[0]/
                 curr_spectra_counts["counts"].sum())
    
ratios = pd.DataFrame({"filtering":["Before filtering","After_filtering"],
                     "b_tnk_ratio":b_tnk_ratio, 
                     "b_all_cells_ratio":b_all_cells_ratio}) 
print(ratios.head())
sns.barplot(data = ratios, x = "filtering", y = "b_all_cells_ratio", ax = ax[-1])
plt.tight_layout()
```

```python
def get_average_spectra_numbers(data):
    cells = data["cell_type"].unique()
    df = pd.DataFrame()
    cell_types = list()
    patient_names = list()
    patient_states = list()
    average_spectra_numbers = list()
    for cell_type in cells: 
        curr_data = data[data["cell_type"]==cell_type]
        for patient_name,patient in curr_data.groupby("patient_name"): 
            average_spectra_number = np.mean([len(np.unique(patient[patient["cell_name"]==cell_name]["spectre"])) 
                                              for cell_name in patient["cell_name"].unique()])
            cell_types.append(cell_type)
            patient_names.append(patient_name)
            patient_states.append(patient["patient_state"].unique()[0])
            average_spectra_numbers.append(average_spectra_number)

    df = pd.DataFrame({"cell_types":cell_types, 
                      "patient_names":patient_names,
                       "patient_states":patient_states,
                      "average_spectra_numbers":average_spectra_numbers})
    return df 

average_spectra_numbers_before_filtering = get_average_spectra_numbers(data)
average_spectra_numbers_after_filtering = get_average_spectra_numbers(data[data["PSD_filtered"]==1])

f,ax = plt.subplots(2,1,figsize = (15,10))
sns.barplot(data = average_spectra_numbers_before_filtering, x = "patient_names", y = "average_spectra_numbers", hue = "cell_types", ax = ax[0])
ax[0].tick_params(axis = "x", rotation = 45)
sns.barplot(data = average_spectra_numbers_after_filtering, x = "patient_names", y = "average_spectra_numbers", hue = "cell_types", ax = ax[1])
ax[1].tick_params(axis = "x", rotation = 45)
plt.tight_layout()
        
    #print([np.unique(patient["spectre"])for patient_name,patient in curr_data.groupby("patient_name"): 
        
```

```python
before_filtering_ratios = []
after_filtering_ratios = []
patient_names = []
patient_states = []

for idp,patient in data.groupby("patient_name"): 
    after_filtering = patient[patient["PSD_filtered"]==1]
    before_filtering_ratio = Counter(patient["cell_type"])["B"]/sum(Counter(patient["cell_type"]).values())
    after_filtering_ratio = Counter(after_filtering["cell_type"])["B"]/sum(Counter(after_filtering["cell_type"]).values())
    before_filtering_ratios.append(before_filtering_ratio)
    after_filtering_ratios.append(after_filtering_ratio)
    patient_names.append(idp)
    patient_states.append(patient["patient_state"].unique()[0])
    

df = pd.DataFrame({"patient_name":patient_names, 
                  'patient_state':patient_states, 
                  "before_filtering_ratio":before_filtering_ratios, 
                  "after_filtering_ratio":after_filtering_ratios})

df["delta"] = df["after_filtering_ratio"]-df["before_filtering_ratio"]
print(df)
f,ax = plt.subplots(figsize = (17,5))
sns.scatterplot(data = df, x = "patient_name", y = "delta", hue = "patient_state", ax = ax, s = 50)
ax.tick_params(axis = "x", rotation = 45)
```

```python
def get_table(data,ratio = False):

    before = data.groupby("cell_type").count()["patient_name"]
    after = data[data["PSD_filtered"]==1].groupby("cell_type").count()["patient_name"]
    spectra_table = pd.concat([before,after],axis = 1)
    spectra_table.columns = ["Before","After"]
    
    return spectra_table.fillna(int(0))

def make_fisher_exact_test(table): 
    return scipy.stats.fisher_exact(table) if table.shape==(2,2) else None

def get_fisher_exact_test_p_values(data):
    p_values = [get_table(data[data["patient_name"]==patient_name])
                if get_table(data[data["patient_name"]==patient_name]).shape==(2,2) else None
                for patient_name in np.unique(data["patient_name"])]
    p_values = pd.DataFrame(p_values,columns = ["statistic","p_value"])
    p_values["patient_name"] = data["patient_name"].unique()
    return p_values

```
